```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr = cv2.imread(image_path)

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# 1. HISTOGRAM EQUALIZATION

hist_eq = cv2.equalizeHist(gray)

# 2. CLAHE (Adaptive Histogram Equalization)

clahe = cv2.createCLAHE(
    clipLimit=2.0,
    tileGridSize=(8,8)
)

clahe_img = clahe.apply(gray)

# 3. CONTRAST ENHANCEMENT

contrast = cv2.convertScaleAbs(
    img_rgb,
    alpha=1.8,   # Contrast control
    beta=0       # Brightness control
)

# 4. BRIGHTNESS ENHANCEMENT

bright = cv2.convertScaleAbs(
    img_rgb,
    alpha=1.0,
    beta=50
)

# 5. GAMMA CORRECTION

gamma = 1.5

gamma_corrected = np.array(
    255 * (img_rgb / 255) ** gamma,
    dtype='uint8'
)

# 6. GAUSSIAN BLUR

gaussian_blur = cv2.GaussianBlur(
    img_rgb,
    (9,9),
    0
)

# 7. MEDIAN BLUR

median_blur = cv2.medianBlur(
    img_rgb,
    5
)

# 8. SHARPENING

kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

sharp = cv2.filter2D(
    img_rgb,
    -1,
    kernel
)

# 9. EDGE DETECTION

edges = cv2.Canny(
    gray,
    100,
    200
)

# 10. THRESHOLDING

_, binary_thresh = cv2.threshold(
    gray,
    127,
    255,
    cv2.THRESH_BINARY
)

# 11. ADAPTIVE THRESHOLDING

adaptive_thresh = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)

# 12. NOISE ADDITION

noise = np.random.normal(
    0,
    25,
    img_rgb.shape
)

noisy_img = np.clip(
    img_rgb + noise,
    0,
    255
).astype(np.uint8)

# 13. DENOISING

denoised = cv2.fastNlMeansDenoisingColored(
    noisy_img,
    None,
    10,
    10,
    7,
    21
)

# VISUALIZATION

fig, ax = plt.subplots(4, 4, figsize=(22,22))

# Row 1
ax[0,0].imshow(img_rgb)
ax[0,0].set_title("Original RGB")

ax[0,1].imshow(gray, cmap='gray')
ax[0,1].set_title("Grayscale")

ax[0,2].imshow(hist_eq, cmap='gray')
ax[0,2].set_title("Histogram Equalization")

ax[0,3].imshow(clahe_img, cmap='gray')
ax[0,3].set_title("CLAHE")

# Row 2
ax[1,0].imshow(contrast)
ax[1,0].set_title("Contrast Enhancement")

ax[1,1].imshow(bright)
ax[1,1].set_title("Brightness Enhancement")

ax[1,2].imshow(gamma_corrected)
ax[1,2].set_title("Gamma Correction")

ax[1,3].imshow(gaussian_blur)
ax[1,3].set_title("Gaussian Blur")

# Row 3
ax[2,0].imshow(median_blur)
ax[2,0].set_title("Median Blur")

ax[2,1].imshow(sharp)
ax[2,1].set_title("Sharpening")

ax[2,2].imshow(edges, cmap='gray')
ax[2,2].set_title("Edge Detection")

ax[2,3].imshow(binary_thresh, cmap='gray')
ax[2,3].set_title("Binary Threshold")

# Row 4
ax[3,0].imshow(adaptive_thresh, cmap='gray')
ax[3,0].set_title("Adaptive Threshold")

ax[3,1].imshow(noisy_img)
ax[3,1].set_title("Noisy Image")

ax[3,2].imshow(denoised)
ax[3,2].set_title("Denoised Image")

# RGB Histogram
colors = ['red', 'green', 'blue']

for i, color in enumerate(colors):
    hist = cv2.calcHist(
        [img_rgb],
        [i],
        None,
        [256],
        [0,256]
    )
    
    ax[3,3].plot(hist, color=color)

ax[3,3].set_title("RGB Histogram")

# Remove Axes
for a in ax.ravel():
    if a != ax[3,3]:
        a.axis('off')

plt.tight_layout()
plt.show()
```
